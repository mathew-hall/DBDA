```r
prior <- c(a=20,b=20)
likelihood <- c(a=17,b=3)
posterior <- prior + likelihood

dx <- 0.001
x <- seq(0,1,dx)
pr.d <- dbeta(x, prior[1], prior[2])
l.d <- dbeta(x, likelihood[1], likelihood[2])
po.d <- dbeta(x, posterior[1], posterior[2])

par(mfrow=c(3,1))
plot(x,pr.d)
plot(x,l.d)
plot(x,po.d)
```
